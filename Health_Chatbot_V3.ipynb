### Actions File

```python
%%writefile actions.py

from typing import Any, Text, Dict, List, Optional, Tuple
import os
import re
import json
from datetime import datetime
import requests

from rasa_sdk import Action, Tracker, FormValidationAction
from rasa_sdk.executor import CollectingDispatcher
from rasa_sdk.types import DomainDict
from rasa_sdk.events import SlotSet, EventType

EMERGENCY_KEYWORDS = {
    "chest pain", "shortness of breath", "severe bleeding", "stroke",
    "loss of consciousness", "fainting", "confusion", "blue lips",
    "severe allergic reaction", "anaphylaxis"
}

SEVERE_KEYWORDS = {
    "high fever", "fever > 39", "uncontrolled vomiting", "persistent severe headache",
    "stiff neck", "blood in stool", "dehydration", "severe abdominal pain"
}

MILD_KEYWORDS = {
    "mild headache", "runny nose", "sore throat", "mild cough", "minor rash",
    "mild nausea", "seasonal allergies"
}

def extract_numbers(text: str) -> Optional[float]:
    m = re.search(r"\d+(\.\d+)?", text or "")
    return float(m.group(0)) if m else None

def normalize_lower_list(maybe_list: Any) -> List[str]:
    if not maybe_list:
        return []
    if isinstance(maybe_list, list):
        return [str(x).lower() for x in maybe_list]
    return [str(maybe_list).lower()]

class ValidateTriageForm(FormValidationAction):
    def name(self) -> Text:
        return "validate_triage_form"

    def validate_symptoms_list(
        self,
        value: Any,
        dispatcher: CollectingDispatcher,
        tracker: Tracker,
        domain: DomainDict,
    ) -> Dict[Text, Any]:
        items = value if isinstance(value, list) else [value]
        items = [str(x).strip() for x in items if str(x).strip()]
        if not items:
            dispatcher.utter_message(text="Please share at least one symptom.")
            return {"symptoms_list": None}
        return {"symptoms_list": items}

    def validate_severity(
        self,
        value: Any,
        dispatcher: CollectingDispatcher,
        tracker: Tracker,
        domain: DomainDict,
    ) -> Dict[Text, Any]:
        allowed = {"mild", "moderate", "severe"}
        val = str(value).lower().strip()
        if val not in allowed:
            dispatcher.utter_message(text="Please choose mild, moderate, or severe.")
            return {"severity": None}
        return {"severity": val}

    def validate_age(
        self,
        value: Any,
        dispatcher: CollectingDispatcher,
        tracker: Tracker,
        domain: DomainDict,
    ) -> Dict[Text, Any]:
        # Accept plain number or text containing a number
        age_num = value if isinstance(value, (int, float)) else extract_numbers(str(value))
        if age_num is None or age_num <= 0 or age_num > 120:
            dispatcher.utter_message(text="Please provide an age between 1 and 120.")
            return {"age": None}
        return {"age": age_num}

    def validate_duration(
        self,
        value: Any,
        dispatcher: CollectingDispatcher,
        tracker: Tracker,
        domain: DomainDict,
    ) -> Dict[Text, Any]:
        if not value or len(str(value).strip()) < 2:
            dispatcher.utter_message(text="Please share how long you've had the symptoms.")
            return {"duration": None}
        return {"duration": value}

class ActionTriage(Action):
    def name(self) -> Text:
        return "triage_action"

    def run(
        self, dispatcher: CollectingDispatcher, tracker: Tracker, domain: DomainDict
    ) -> List[EventType]:
        symptoms = normalize_lower_list(tracker.get_slot("symptoms_list"))
        severity = (tracker.get_slot("severity") or "unknown").lower()
        age = tracker.get_slot("age") or 0
        duration = (tracker.get_slot("duration") or "").lower()

        # Basic rule-based triage
        text_blob = " ".join(symptoms + [severity, duration])

        level = "unknown"
        explanation = []

        # Emergency checks
        if any(k in text_blob for k in EMERGENCY_KEYWORDS):
            level = "emergency"
            explanation.append("Reported red-flag symptoms that may be life-threatening.")

        # Severe logic
        elif severity == "severe" or any(k in text_blob for k in SEVERE_KEYWORDS):
            level = "severe"
            explanation.append("Symptoms or severity suggest prompt clinical evaluation.")

        # Moderate logic
        elif severity == "moderate" or (age < 2 or age > 70 and "fever" in text_blob):
            level = "moderate"
            explanation.append("Symptoms warrant medical assessment, especially given age or persistence.")

        # Mild logic
        elif severity == "mild" or any(k in text_blob for k in MILD_KEYWORDS):
            level = "mild"
            explanation.append("Symptoms sound mild and may improve with self-care.")

        # Duration-based bump
        if level in {"mild", "unknown"} and any(
            kw in duration for kw in ["week", "weeks", "10 days", "two weeks", "> 7 days"]
        ):
            level = "moderate"
            explanation.append("Duration is prolonged, recommending evaluation.")

        # Pediatric caution
        if age and age < 2 and "fever" in text_blob and level != "emergency":
            level = "moderate"
            explanation.append("Fever in very young children needs clinician review.")

        explanation_text = " ".join(explanation) if explanation else "General guidance provided."

        # Send result messages
        if level == "emergency":
            dispatcher.utter_message(text="Your symptoms may be urgent. Please seek emergency care now.")
        elif level == "severe":
            dispatcher.utter_message(response="utter_triage_result_severe")
        elif level == "moderate":
            dispatcher.utter_message(response="utter_triage_result_moderate")
        else:
            dispatcher.utter_message(response="utter_triage_result_mild")

        return [SlotSet("triage_level", level), SlotSet("triage_explanation", explanation_text)]

class ValidateAppointmentForm(FormValidationAction):
    def name(self) -> Text:
        return "validate_appointment_form"

    def validate_doctor_specialty(
        self, value: Any, dispatcher: CollectingDispatcher, tracker: Tracker, domain: DomainDict
    ) -> Dict[Text, Any]:
        if not value or len(str(value).strip()) < 3:
            dispatcher.utter_message(text="Please provide a valid specialty, like 'dermatology'.")
            return {"doctor_specialty": None}
        return {"doctor_specialty": str(value).lower()}

    def validate_doctor_name(
        self, value: Any, dispatcher: CollectingDispatcher, tracker: Tracker, domain: DomainDict
    ) -> Dict[Text, Any]:
        val = str(value).strip().lower()
        if val in {"", "no", "no preference", "any"}:
            return {"doctor_name": "no preference"}
        return {"doctor_name": value}

    def validate_appointment_date(
        self, value: Any, dispatcher: CollectingDispatcher, tracker: Tracker, domain: DomainDict
    ) -> Dict[Text, Any]:
        val = str(value).strip().lower()
        if val in {"asap", "as soon as possible", "soonest"}:
            return {"appointment_date": "ASAP"}
        try:
            datetime.strptime(val, "%Y-%m-%d")
            return {"appointment_date": val}
        except Exception:
            dispatcher.utter_message(text="Please use YYYY-MM-DD or say 'as soon as possible'.")
            return {"appointment_date": None}

    def validate_time_slot(
        self, value: Any, dispatcher: CollectingDispatcher, tracker: Tracker, domain: DomainDict
    ) -> Dict[Text, Any]:
        if not value or len(str(value).strip()) < 3:
            dispatcher.utter_message(text="Please share a preferred time window, like '10:00' or 'afternoon'.")
            return {"time_slot": None}
        return {"time_slot": value}

    def validate_patient_name(
        self, value: Any, dispatcher: CollectingDispatcher, tracker: Tracker, domain: DomainDict
    ) -> Dict[Text, Any]:
        if not value or len(str(value).split()) < 1:
            dispatcher.utter_message(text="Please provide the name for the booking.")
            return {"patient_name": None}
        return {"patient_name": value}

    def validate_contact_phone(
        self, value: Any, dispatcher: CollectingDispatcher, tracker: Tracker, domain: DomainDict
    ) -> Dict[Text, Any]:
        digits = re.sub(r"\D", "", str(value))
        if len(digits) < 7:
            dispatcher.utter_message(text="Please provide a valid phone number.")
            return {"contact_phone": None}
        return {"contact_phone": value}

    def validate_contact_email(
        self, value: Any, dispatcher: CollectingDispatcher, tracker: Tracker, domain: DomainDict
    ) -> Dict[Text, Any]:
        if not value:
            return {"contact_email": ""}
        pattern = r"^[^\s@]+@[^\s@]+\.[^\s@]+$"
        if not re.match(pattern, str(value)):
            dispatcher.utter_message(text="That email doesn’t look valid. You can leave it blank if you prefer.")
            return {"contact_email": None}
        return {"contact_email": value}

class ActionQueryDoctorDb(Action):
    def name(self) -> Text:
        return "query_doctor_db"

    def run(self, dispatcher: CollectingDispatcher, tracker: Tracker, domain: DomainDict) -> List[EventType]:
        base_url = os.environ.get("API_BASE_URL", "")
        api_key = os.environ.get("API_KEY", "")
        specialty = tracker.get_slot("doctor_specialty") or "general practice"
        doctor_name = tracker.get_slot("doctor_name") or "no preference"
        date = tracker.get_slot("appointment_date") or "ASAP"
        time_pref = tracker.get_slot("time_slot") or ""

        headers = {"Authorization": f"Bearer {api_key}"} if api_key else {}
        params = {
            "specialty": specialty,
            "doctor": None if doctor_name == "no preference" else doctor_name,
            "date": date,
            "time_pref": time_pref,
        }

        try:
            if not base_url:
                # Simulated response for local dev
                simulated = {
                    "doctor": doctor_name if doctor_name != "no preference" else f"Dr. Lee ({specialty})",
                    "slots": [
                        {"date": date if date != "ASAP" else "2025-09-01", "time": "10:30"},
                        {"date": date if date != "ASAP" else "2025-09-01", "time": "14:00"},
                    ],
                }
                dispatcher.utter_message(text=f"I found availability for {simulated['doctor']}.")
                return [
                    SlotSet("recommended_doctor", simulated["doctor"]),
                    SlotSet("available_slots", simulated["slots"]),
                ]

            resp = requests.get(f"{base_url}/availability", headers=headers, params=params, timeout=10)
            resp.raise_for_status()
            data = resp.json()

            doctor = data.get("doctor") or f"Clinician ({specialty})"
            slots = data.get("slots", [])
            if not slots:
                dispatcher.utter_message(text="No slots match your preferences. Would you like different times or another day?")
                return [SlotSet("recommended_doctor", doctor), SlotSet("available_slots", [])]

            dispatcher.utter_message(text=f"I found availability for {doctor}.")
            # Optionally present the first few options
            pretty = ", ".join([f"{s.get('date')} {s.get('time')}" for s in slots[:3]])
            dispatcher.utter_message(text=f"Soonest times: {pretty}")
            return [SlotSet("recommended_doctor", doctor), SlotSet("available_slots", slots)]

        except requests.RequestException:
            dispatcher.utter_message(text="I couldn’t reach the scheduling system right now.")
            return []

class ActionSubmitAppointment(Action):
    def name(self) -> Text:
        return "action_submit_appointment"

    def run(self, dispatcher: CollectingDispatcher, tracker: Tracker, domain: DomainDict) -> List[EventType]:
        base_url = os.environ.get("API_BASE_URL", "")
        api_key = os.environ.get("API_KEY", "")
        doctor = tracker.get_slot("recommended_doctor") or tracker.get_slot("doctor_name") or "Clinician"
        specialty = tracker.get_slot("doctor_specialty") or "general practice"
        date = tracker.get_slot("appointment_date")
        time_pref = tracker.get_slot("time_slot")
        patient_name = tracker.get_slot("patient_name")
        phone = tracker.get_slot("contact_phone")
        email = tracker.get_slot("contact_email") or ""

        payload = {
            "doctor": doctor,
            "specialty": specialty,
            "date": date,
            "time": time_pref,
            "patient_name": patient_name,
            "phone": phone,
            "email": email,
        }
        headers = {"Authorization": f"Bearer {api_key}", "Content-Type": "application/json"} if api_key else {"Content-Type": "application/json"}

        try:
            if not base_url:
                # Simulate booking success
                booking_ref = "BK-" + datetime.now().strftime("%Y%m%d%H%M%S")
                dispatcher.utter_message(response="utter_booking_confirmed")
                return [SlotSet("booking_reference", booking_ref)]
            resp = requests.post(f"{base_url}/book", headers=headers, data=json.dumps(payload), timeout=10)
            resp.raise_for_status()
            data = resp.json()
            booking_ref = data.get("reference") or "BK-LOCAL"
            return [SlotSet("booking_reference", booking_ref)]
        except requests.RequestException:
            dispatcher.utter_message(response="utter_booking_failed")
            return []

class ActionHandoffToHuman(Action):
    def name(self) -> Text:
        return "action_handoff_to_human"

    def run(self, dispatcher: CollectingDispatcher, tracker: Tracker, domain: DomainDict) -> List[EventType]:
        dispatcher.utter_message(text="Transferring you to a human agent. If this is urgent, please call emergency services.")
        return []

class ActionDefaultFallback(Action):
    def name(self) -> Text:
        return "action_default_fallback"

    def run(self, dispatcher: CollectingDispatcher, tracker: Tracker, domain: DomainDict) -> List[EventType]:
        dispatcher.utter_message(text="I’m not sure I understood. I can help with symptoms, triage, or bookings.")
        dispatcher.utter_message(response="utter_disclaimer_short")
        return []

#Insurance

#class ActionCheckInsurance(Action):
#    def name(self) -> Text:
#        return "action_check_insurance"
#
#    def run(self, dispatcher: CollectingDispatcher, tracker: Tracker, domain: DomainDict) -> List[EventType]:
#        insurance = tracker.get_slot("insurance")
#        if not insurance:
#            dispatcher.utter_message(response="utter_insurance")
#            return []
#
#        insurance = str(insurance).lower().strip()
#        if insurance in ACCEPTED_INSURANCES:
#            status = "accept"
#            additional_info = "Congrats, your insurance is accepted in our clinic."
#        else:
#            status = "do not accept"
#            additional_info = "Unfortunately we don't accept this insurance but we have many options of payment."

#        dispatcher.utter_message(
#            response="utter_insurance_result",
#            insurance=insurance,
#            insurance_status=status,
#            additional_info=additional_info
#        )
#        return [SlotSet("insurance_status", status)]
```

### App File

```python
import streamlit as st
import requests

# --- CONFIG ---
RASA_REST_API_URL = "http://localhost:5005/webhooks/rest/webhook"  # Adjust if hosted elsewhere

st.set_page_config(page_title="Clinic Assistant", page_icon="💬")

# --- INIT SESSION STATE ---
if "messages" not in st.session_state:
    st.session_state.messages = [
        {"role": "assistant", "content": "Hi 👋 I’m your clinic assistant. I’m not a doctor; this isn’t medical advice. How can I help today?"}
    ]

# --- SIDEBAR INFO ---
st.sidebar.title("About this chatbot")
st.sidebar.markdown(
    """
    **Phase 1**: Handles FAQs and symptom triage only.
    • Clinic hours, location or services
    • Symptom intake and advice
    • Emergency escalation and human handoff
    """
)

# --- MAIN CHAT LOOP ---
st.title("🏥 Clinic FAQ & Triage Bot")

# Display chat history
for msg in st.session_state.messages:
    if msg["role"] == "user":
        st.markdown(f"**You:** {msg['content']}")
    else:
        st.markdown(f"**Bot:** {msg['content']}")

# User input form
with st.form(key="chat_form", clear_on_submit=True):
    user_input = st.text_input("Your message:", "")
    submitted = st.form_submit_button("Send")

if submitted and user_input.strip():
    # Append user message
    st.session_state.messages.append({"role": "user", "content": user_input})

    try:
        # Send to Rasa REST API
        response = requests.post(
            RASA_REST_API_URL,
            json={"sender": "streamlit_user", "message": user_input},
            timeout=10
        )
        response.raise_for_status()

        bot_messages = response.json()
        if bot_messages:
            for bot_msg in bot_messages:
                text = bot_msg.get("text", "")
                if text:
                    st.session_state.messages.append({"role": "assistant", "content": text})
        else:
            st.session_state.messages.append({"role": "assistant", "content": "🤔 I didn’t get a response from the server."})

    except requests.RequestException:
        st.session_state.messages.append(
            {"role": "assistant", "content": "⚠️ Could not reach the chatbot server. Please check the Rasa service."}
        )

    st.experimental_rerun()
```
